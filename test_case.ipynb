```python
users = pd.read_csv(os.path.join('data/users_test.csv'), delimiter=',')
conns = pd.read_csv(os.path.join('data/conns_test.csv'), delimiter=',')
hotspots = pd.read_csv(os.path.join('data/hotspots_test.csv'), delimiter=',')
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
user_id = 15505
user_hotspots = hotspots[hotspots['owner_id'] == user_id]
user_hotspots
```

```python
user_hotspots['created_at'] = user_hotspots['created_at'].apply(lambda x: pd.to_datetime(x).date())
user_hotspots['created_at_month'] = user_hotspots['created_at'].apply(lambda x: x.strftime('%Y-%m'))
```

## Количество созданых хотспотов по времени

```python
fix, ax = plt.subplots(figsize=(20, 5))
plt.xticks(rotation=90)

sns.lineplot(
    x='created_at_month',
    y='id',
    data=user_hotspots.groupby(by='created_at_month')['id'].count().reset_index().sort_values(by='created_at_month'),
    ax=ax
)
```

## Количество хотспотов пользователя по странам

```python
fix, ax = plt.subplots(figsize=(20, 5))
plt.xticks(rotation=90)

sns.lineplot(
    x='country_code',
    y='id',
    data=user_hotspots.groupby(by='country_code')['id'].count().reset_index().sort_values(by='id'),
    ax=ax
)
```

```python
user_conns = conns[conns['hotspot_id'].isin(user_hotspots['id'].values)]
user_conns
```

## Распределение среднего качества связи хотспотов пользователя

```python
user_conns.groupby(by='hotspot_id')['signal_strength'].mean().reset_index()['signal_strength'].plot.box()
```

### Количество актуальных и удаленных хотспотов пользователя True = актуальный, False - удаленный

```python
user_hotspots['deleted_at'].isnull().value_counts().plot.bar()
```

### Распределение "вероятности" наличия интернета на хотспотах пользователя

```python
user_conns.groupby('hotspot_id')['is_internet_available'].mean().plot.box()
```
